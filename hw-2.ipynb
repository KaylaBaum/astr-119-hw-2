Import numpy and matplotlib

```python
#%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

In a single line, use numpy to create an array x running x = [0,2*pi] inclusive with 1000 values.

```python
x = np.linspace(0,2*np.pi,1000)
```

Use matplotlib to plot the following functions on a single plot using the x range x = [0,2*pi] and the vertical range y = [-1,10]:

    y(x) = 5.5 cos(2 * x) + 5.5

    y(x) = 0.02 * exp(x)

    y(x) = 0.25 * x^2 + 0.1 sin (10 * x)

Label the y axis "Measures of Awesomeness" and the x axis "Time in ASTR / EART 119".

```python
y = (5.5 * np.cos(2*x) + 5.5)
z = (0.02 * np.exp(x))
w = (0.25 * (x**2.0) + 0.1 * np.sin(10*x))

plt.plot(x, y, 'r-')
plt.plot(x, z, 'b-')
plt.plot(x, w, 'g-')
plt.xlabel('Time in ASTR / EART 119')
plt.ylabel('Measures of Awesomeness')
plt.xlim([0,2*np.pi])
plt.ylim([-1,10])
```
